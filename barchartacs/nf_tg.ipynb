### Compare UK and Dutch Natural Gas

```python
import pandas as pd
import datetime
import plotly_utilities as pu
from IPython import display
```

```python
def run_analysis(mc):
    nf = f'nf{mc}22_price-history.csv'
    tg = f'tg{mc}22_price-history.csv'
    gbp = '^gbpusd_price-history.csv'
    eur = '^eurusd_price-history.csv'
    cmd_dict = {}
    df_all = None
    for fn in [nf,tg,gbp,eur]:
        sym = fn[0:2] if fn[0]!='^' else fn[1:4]
        df = pd.read_csv(f'temp_folder/{fn}').rename(columns={'Last':sym})[['Time',sym]].iloc[0:200]
        df['Time'] = [
            datetime.datetime.strptime(d,'%m/%d/%Y') for d in df['Time'].astype(str).values
        ]
        cmd_dict[sym] = df.copy()
        if df_all is None:
            df_all = df.copy()
        else:
            df_all = df_all.merge(df,on='Time',how='left')
    df_all['Time'] = [d.year*100*100 + d.month*100 + d.day for d in df_all['Time']]
    df_all = df_all.sort_values('Time')
    df_all.index = range(len(df_all))
    df_all = df_all.dropna()
    df_all['nf_mmbtu'] = df_all.nf * df_all.gbp / 10 
    df_all['tg_mmbtu'] = df_all.tg * df_all.eur / 3.412
    df_all['nf_tg_mmbtu'] = df_all.nf_mmbtu - df_all.tg_mmbtu  
    return df_all
```

```python
df = run_analysis('z')
display.display(df.tail(20))
display.display(pu.plotly_plot(df.iloc[0:-10][['Time','nf_tg_mmbtu']],x_column='Time'))
```

```python
df = run_analysis('u')
display.display(df[(df.Time>20220601) & (df.Time<20220801)])
display.display(pu.plotly_plot(df.iloc[0:-1][['Time','nf_tg_mmbtu']],x_column='Time'))
```
